```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("/content/Screentime-App-Details.csv")
```

```python
df.head()
```

```python
# check null values
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
figure = px.bar(data_frame=df, 
                x = "Date", 
                y = "Usage", 
                color="App", 
                title="Usage")
figure.show()
```

```python
figure = px.bar(data_frame=df,
                x = "Date",
                y = "Times opened",
                color = "App",
                title = "Times opened" )
figure.show()
```

```python
figure = px.bar(data_frame=df,
                x = "Date",
                y = "Notifications",
                color = "App",
                title = "Notifications" )
figure.show()
```

```python
figure = px.bar(data_frame=df, 
                x = "Date", 
                y = "Notifications", 
                color="App", 
                title="Notifications")
figure.show()
```

```python
figure = px.scatter(data_frame = df, 
                    x="Notifications",
                    y="Usage", 
                    size="Notifications", 
                    trendline="ols", 
                    title = "Relationship Between Number of Notifications and Usage")
figure.show()
```
