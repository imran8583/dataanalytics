## **Problem Statements:**


### **Question 1: What are the types for aviation facilities in Alaska?**

In this dataset, there is a variable called Type that records the type of a specific aviation facility. To answer this question, we would need to know what are the unique entries for this variable.

So naturally, we'll need to first extract this column from the entire dataset. The way we extract a column from a dataframe is similar to how we extract a column from a Python dictionary, except that we pass in the column (variable) name instead of key value:

```python
df_facilities["Type"].head(10)
```

```python
df_facilities["Type"].unique()
```

Alaska has 3 types of avation facilities that is 'AIRPORT', 'SEAPLANE BASE' and 'HELIPORT'.

### **Question 2: Provide a table containing the information for all Seaplane Bases in Alaska**

```python
seaplane_base = df_facilities[df_facilities["Type"] == 'SEAPLANE BASE']

print("Table containing all Seaplane Bases in Alaska")
print(f"Alaska has a total of {len(seaplane_base)}  Seaplane Bases")
seaplane_base.head()
```

### **Question 3: What are the 5 highest aviation facilities in Alaska?**

In this dataset, the ARPElevation variable records the elevation of aviation facilities in Alaska in feet. What we will do is to sort the rows of this dataset in reverse based on the value of column ARPElevation. This can be done with sort_values method of Pandas dataframe:

```python
df_facilities["ARPElevation"]
```

```python
top5_aviation = df_facilities.sort_values(by = "ARPElevation", ascending = False).head()
```

```python
top5_aviation
```

### **Question 4: What is the average elevation of aviation facilities?**

We we already saw in this dataset, the ARPElevation tells us the elevation of aviation facilities in Alaska. So we can use it to find the mean or average

```python
avg_elevation_aviation = df_facilities["ARPElevation"].mean()
avg_elevation_aviation
```

```python
print(f"As we can see from the above result, the average elevation of aviation facilities is almost {round(avg_elevation_aviation)} ft.")
```

### **Question 5: What are the highest and lowest elevation values of the aviation facilities?**

 Well, similar to the previous question, Pandas has built-in functions that help us find the maximum (largest) and minimum (lowest) values in a column. These functions are max() and min(), respectively.

```python
highest_elevation_values = df_facilities['ARPElevation'].max()
lowest_elevation_values = df_facilities['ARPElevation'].min()

print(f"The highest elevation values of the aviation facilities {highest_elevation_values} ft.")
print(f"The lowest elevation values of the aviation facilities {lowest_elevation_values} ft.")
```
