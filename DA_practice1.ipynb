# **Display random numbers plot**

```python
# import all the neccessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
# define or import the required dataset

randomNum = np.random.rand(10)
randomNum
```

```python
# set the plot parameters
# display the created plot

style.use("ggplot")

plt.plot(randomNum, "g", label="line one", linewidth=2)
plt.xlabel("Range")
plt.ylabel("Numbers")
plt.title("random Numbers")
plt.legend()
plt.show()
```

#**Ohio State - 2012 : Leading Causes of Death**

```python
# import all the neccessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
# define or import the required dataset

cause = "Chronic Diseases", "Unintentional Injuries", "Alzhimers", "influenza and Pneumonia", "Sepsis", "Others"

percentile = [62,5,4,2,1,26]
```

```python
# set the plot parameters
# display the created plot

plt.figure(figsize=(9,9))
explode = (0.03,0,0,0,0,0)
plt.pie(percentile, labels=cause, explode=explode,autopct="%1.1f%%", startangle=70)
plt.axis("equal")
plt.title("Ohio State - 2012 : Leading Causes of Death")
plt.legend()
plt.show()
```
